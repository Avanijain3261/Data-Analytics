Implementing different Classifiers and Compare their performance.



```python
# Step-1: Load data set
from sklearn.datasets import load_breast_cancer
x, y = load_breast_cancer(return_X_y=True)
print(x.shape)
print(y.shape)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Load Classifiers
from sklearn.linear_model import Perceptron, LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

clf1 = Perceptron()
clf2 = LogisticRegression()
clf3 = SVC()
clf4 = DecisionTreeClassifier()
clf5 = KNeighborsClassifier()
clf6 = GaussianNB()

classifiers = [clf1, clf2, clf3, clf4, clf5, clf6]
classifiers_name = ['Perceptron', 'LogisticRegression', 'SVC', 'DecisionTreeClassifier', 'KNeighborsClassifier', 'GaussianNB']

acc={}
t={}

import time
from sklearn.metrics import accuracy_score

for clf, clf_name in zip(classifiers, classifiers_name):
    start = time.time()
    clf.fit(x_train, y_train)
    end=time.time()
    y_pred = clf.predict(x_test)
    acc[clf_name]=accuracy_score(y_test, y_pred)
    t[clf_name]=end-start

print('----------------------Accuracy Comparision----------------------')
for i,j in acc.items():
  print(i,':-',j)
print('----------------------Training Time Comparision----------------------')
for i,j in t.items():
  print(i,':-',j)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.bar(acc.keys(), acc.values())
plt.xlabel('Classifiers')
plt.ylabel('Accuracy')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.bar(t.keys(), t.values())
plt.xlabel('Classifiers')
plt.ylabel('Time')
plt.show()
```

```python
## Ensemble Learning
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, BaggingClassifier # Bagging
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier # Boosting
from sklearn.ensemble import VotingClassifier # Stacking

clf7 = RandomForestClassifier()
clf8 = ExtraTreesClassifier()
clf9 = BaggingClassifier()
clf10 = AdaBoostClassifier()
clf11 = GradientBoostingClassifier()
clf12 = VotingClassifier(estimators=[('Perceptron', clf1),('LogisticRegression', clf2), ('SVC', clf3)],
                         voting='hard', weights=[1,1,1])

classifiers = [clf7, clf8, clf9, clf10, clf11, clf12]
classifiers_name = ['RandomForestClassifier', 'ExtraTreesClassifier', 'BaggingClassifier',
                    'AdaBoostClassifier', 'GradientBoostingClassifier', 'VotingClassifier']

acc={}
t={}

import time
from sklearn.metrics import accuracy_score

for clf, clf_name in zip(classifiers, classifiers_name):
    start = time.time()
    clf.fit(x_train, y_train)
    end=time.time()
    y_pred = clf.predict(x_test)
    acc[clf_name]=accuracy_score(y_test, y_pred)
    t[clf_name]=end-start

print('----------------------Accuracy Comparision----------------------')
for i,j in acc.items():
  print(i,':-',j)
print('----------------------Training Time Comparision----------------------')
for i,j in t.items():
  print(i,':-',j)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.bar(acc.keys(), acc.values())
plt.xlabel('Classifiers')
plt.ylabel('Accuracy')
plt.show()
```
