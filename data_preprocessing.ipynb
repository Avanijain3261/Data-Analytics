DATA PRE-PROCESSING

```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/basics_data.csv')
df
```

```python
#Missing Value
df.isna().sum()
```

```python
#Method-1 Delete
df1=df.dropna()
df1
```

```python
df[['F2','F3']] # F2 & F3 column
df.iloc[:,[2,3]] #col with index
```

```python
# Method-2 Imputation
import numpy as np
from sklearn.impute import SimpleImputer
si= SimpleImputer(missing_values=np.nan, strategy='mean') # median, most_frequent
df[['F4','F5']]= si.fit_transform(df[['F4','F5']])
df
```

```python
# Check Categorical Data
df.info()
```

```python
# Method-1 Label Encoder (Target- Male/Female)
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Target']=le.fit_transform(df['Target'])
print(le.classes_)
df
```

```python
# One Hot Encoder
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
df1 = pd.DataFrame(ohe.fit_transform(df[['Color']]).toarray(),
                   columns=['Color_'+str(i) for i in range(len(df['Color'].value_counts()))])
df1
df2 = pd.concat([df,df1], axis=1)
df2
df3=df2.drop(['Color'],axis=1)
df3
```

```python
dfc = pd.DataFrame(ohe.fit_transform(df[['Country']]).toarray(),
                   columns=['Country_'+str(i) for i in range(len(df['Country'].value_counts()))])
dfc
df3 = pd.concat([df3,dfc], axis=1)
df3
df4=df3.drop(['Country'], axis=1)
df4
```

```python
df4.info()
```

```python
# Input and Target Split
y=df4['Target']
x=df4.drop(['Target'],axis=1)
print(x.shape)
print(y.shape)
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Feature Scaling, Standardization
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_std=sc.fit_transform(x_train)
x_test_std=sc.transform(x_test)
print(x_test_std)
```

```python
# Normalization
from sklearn.preprocessing import MinMaxScaler
mn=MinMaxScaler()
x_train_norm=mn.fit_transform(x_train)
x_test_norm=mn.transform(x_test)
print(x_test_norm)
```

```python
# Feature Extraction
# PCA
from sklearn.decomposition import PCA
pca=PCA(n_components=5) # <= No. of Features
x_train_pca=pca.fit_transform(x_train) # Unsupervised
x_test_pca=pca.transform(x_test)
print(x_test_pca)
```

```python
#LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda=LDA(n_components=1) # min(1, No. of Class - 1)
x_train_lda=lda.fit_transform(x_train,y_train) # Supervised
x_test_lda=lda.transform(x_test)
print(x_test_lda)
```
